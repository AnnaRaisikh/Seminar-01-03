```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import seaborn as sns
import re
import io
import scipy.stats as stats
```

# 1.3 output equation #1

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/Labour.csv')
df
```

```python
model=smf.ols(data=df, formula='output~capital+labour+wage').fit()
hat_y=model.predict()
```

```python
hat_y
```

```python
df['out2']=hat_y**2
df['out3']=hat_y**3
df['out4']=hat_y**4
df
```

```python
model2=smf.ols(data=df, formula='output~capital+labour+wage+out2+out3+out4').fit()
```

```python
R2_un=model2.rsquared
R2_re=model.rsquared
```

```python
F_obs=(R2_un-R2_re)/(1-R2_un)*(len(df)-7)/3
F_obs
```

```python
alpha=0.05
stats.f.ppf(1-alpha,3,len(df)-7)
```

```python
$$H_0: \gamma_1=\gamma_2=\gamma_3=0$$
$$H_1: \gamma_1^2+\gamma_2^2+\gamma_3^2>0$$
```

# 1.2 sleep equation #2

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/sleep75.csv')
df
```

```python
df['totwrk*south']=df['totwrk']*df['south']
df
```

```python
model1=smf.ols(data=df, formula='sleep~totwrk+south+totwrk*south+age+age**2+male+smsa+marr').fit()
ht_y=model1.predict()
ht_y
```

```python
df['slp2']=ht_y**2
df['slp3']=ht_y**3
df['slp4']=ht_y**4
df
```

```python
model2=smf.ols(data=df, formula='sleep~totwrk+south+totwrk*south+age+age**2+male+smsa+marr+slp2+slp3+slp4').fit()
```

```python
R2_un=model2.rsquared
R2_re=model1.rsquared
```

```python
F_obs=(R2_un-R2_re)/(1-R2_un)*(len(df)-12)/3
F_obs
```

```python
alpha=0.05
stats.f.ppf(1-alpha,3,len(df)-12)
```

$F_{cr}>F_{obs}$ =>H0 accept
